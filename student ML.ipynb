#1. Start by importing all necessary modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mtp
from sklearn import metrics
```

#2. Import dataset

```python
data = pd.read_csv(r'D:\DataScience\my data\students.csv')
df=pd.DataFrame(data)
print(df.to_string())
```

#Conversion of output from good/bad to 1/0 for easier calculations

```python
df['Target'] = df['Target'].map({'Dropout': 0, 'Enrolled': 1, 'Graduate': 2})
```

#3. Select features for x and y

```python
x =df.iloc[:, :-1].values # independent variable
y =df.iloc[:, -1].values # dependant variable
```

#4. Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

#5. Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

#6. Fitting Decision Tree classifier to the training set

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(solver='lbfgs', max_iter=10000)
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)
df2=pd.DataFrame(x_test)
print(df2.to_string)
```

#7.From sklearn.tree import DecisionTreeClassifier

#8. Evaluating Algorithm

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
acc1=metrics.accuracy_score(y_test, y_pred)*100
print("Accuracy:",acc1)
```
